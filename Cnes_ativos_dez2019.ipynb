```python
brasil = pd.read_csv('brazil-population-1998-2017.csv', sep=';', encoding='utf8')
brasil.head()
```

```python
brasil17 = brasil[brasil['year'] == 2017]
brasil17.head()
```

```python
brasil17.rename(columns={'code': 'co_ibge'}, inplace=True)
brasil17.head() 
```

```python
brasil17.drop(['state','abbreviation','lat','long','year', 'population'], axis=1).head()
```

```python
df = cnes.join(brasil17.set_index('co_ibge'), on='co_ibge')
```

```python
df.head()
```

```python
df.drop(['municipio', 'abbreviation', 'year', 'population'], axis=1, inplace=True)
```
